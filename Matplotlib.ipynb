# Multiline Graph


```python
no=range(0,9)
plt.plot([i**2 for i in no],'r')
plt.plot([i**3 for i in no],'g')
plt.plot([i**4 for i in no],'b')
```

```python
plt.plot([i**2 for i in range(0,9)],'r-.',[i**3 for i in range(0,9)],'b--',[i**4 for i in range(0,9)],'cv:')
```

```python
plt.plot([i**2 for i in np.arange(0,9)],'ko',[i**3 for i in np.arange(0,9)],'b--',[i**4 for i in np.arange(0,9)],'cv:')
```

```python
def walk():
    x=np.random.randn(50)
    return np.cumsum(x)


X=walk()
Y=walk()
Z=walk()
plt.plot(X)
plt.plot(Y)
plt.plot(Z)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.plot(X)
plt.plot(Y)
plt.plot(Z)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title',fontsize=10)
plt.suptitle('GRAPH', fontsize=16, fontweight='bold')
```

```python
plt.plot(X,label='X')
plt.plot(Y,label='Y')
plt.plot(Z,label='Z')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title',fontsize=10)
plt.suptitle('GRAPH', fontsize=16, fontweight='bold')
plt.legend(loc=3)
```
