```python
def walk():
    x=np.random.randn(50)
    return np.cumsum(x)


X=walk()
Y=walk()
Z=walk()
plt.plot(X)
plt.plot(Y)
plt.plot(Z)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.plot(X)
plt.plot(Y)
plt.plot(Z)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title',fontsize=10)
plt.suptitle('GRAPH', fontsize=16, fontweight='bold')
```

```python
plt.plot(X,'r-o',label='X')
plt.plot(Y,'g-o',label='Y')
plt.plot(Z,'b-o',label='Z')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title',fontsize=10)
plt.suptitle('GRAPH', fontsize=16, fontweight='bold')
plt.legend(loc=3)
```

# Bar Graph

```python
objects = ('Python', 'Java Script', 'R', 'Ruby', 'Scala', 'Java')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')

plt.show()
```

```python
objects = ('Python', 'Java Script', 'R', 'Ruby', 'Scala', 'Java')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]

plt.barh(y_pos, performance, align='center', alpha=0.5) #barh for horizontal
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')

plt.show()
```

# Error bar charts

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
e = [0.5, 1., 1.5, 2.]
plt.errorbar(x, y, yerr=e, fmt='o')
plt.show()
```

```python
x = np.arange(0, 4, 0.2)
y = np.exp(-x)
e1 = 0.1 * np.abs(np.random.randn(len(y)))
plt.errorbar(x, y, yerr=e1, fmt='.-');
plt.show()
```

# Pie Charts
